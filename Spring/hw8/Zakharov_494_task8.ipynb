```python
import numpy as np
import pylab as plt
```

```python

'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    F = np.asmatrix(F)
    Sigma = 1.0 / l * F * np.transpose(F)
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(Sigma)

    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / \
                dTot   for i in range(n)])
    
    m = np.searchsorted(var_i, varRetained)
    print '{:.2f} variance retained in {} dimensions'.format(var_i[m], m + 1)

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = U[:, :(m + 1)]
    G = (F.T * U_reduced).T

    return G, U_reduced
```

```python
def plot_image(image_id, dataset=X, im_size=28):
    plt.imshow(255-dataset[image_id,:].reshape(im_size, im_size), cmap='gray')
    
def plot_grid(image_ids, nrows, ncols, dataset=X, im_size=8):
    fig = pyplot.gcf()
    fig.set_size_inches(17.5,5)
    for pylab_index, example_id in enumerate(image_ids):
        plt.subplot(nrows, ncols, pylab_index + 1)
        plot_image(example_id, dataset, im_size, )
        plt.axis('off')
    plt.show()
    
plot_grid(range(10), nrows=1, ncols=10)
```

```python
tsne = TSNE(n_components=2, perplexity=40, method='exact')
X_new = tsne.fit_transform(X)
```

```python
plt.figure(figsize=(14,7))
plt.scatter(X_new[:,0], X_new[:,1], c=y)
```

```python
min_met = np.inf
eps = None
min_samples = None
```

```python
for eps_, min_samples_ in itertools.product(np.arange(2, 3, 0.01), xrange(1, 50)):
    db = DBSCAN(eps=eps_, min_samples=min_samples_) 
    clustered = db.fit(X_new)
    labels = clustered.labels_
    if len(set(labels)) == 11:
        if metrics.calinski_harabaz_score(X_new, labels) < min_met:
            eps = eps_
            min_samples = min_samples_
```

```python
print "eps = {:.2f}, min samples = {}".format(eps, min_samples)
```

```python
db = DBSCAN(eps=eps, min_samples=min_samples) 
clustered = db.fit(X_new)
labels = clustered.labels_
```
